```python
def NL2SQL(text):

    words_list = text.lower().split(' ')

    result_df = pd.DataFrame(columns = cal_df.columns[1:-1])
    result_df_cols_dict = {x:x.split('/') for x in result_df.columns}

    query_df_cols = [x.split('/') for x in cal_df.columns[1:-1]]

    values = []

    for i in result_df_cols_dict:
        v = 0
        for j in words_list:
            if j in result_df_cols_dict[i]:
                v += 1
            else:
                v += 0
        if v>0:
            values.append(1)
        else:
            values.append(0)

    sql_value = [
        str(values[0]) + 
        str(values[1]) + 
        str(values[2]) + 
        str(values[3]) + 
        str(values[4]) + 
        str(values[5]) + 
        str(values[6]) + 
        str(values[7]) + 
        str(values[8]) + 
        str(values[9])
             ]

    if sql_value[0] == '1000000000':
        result = mysql('''Select * from df''')
    elif sql_value[0] == '0010000000':
        result = mysql('''Select * from df limit '''+words_list[words_list.index(list(set(words_list)&set(query_df_cols[2]))[0]) - 1])
    elif sql_value[0] == '0001100000':
        result = mysql(q3)
    elif sql_value[0] == '0001000100':
        result = mysql(q4)
    elif sql_value[0] == '0001110000':
        result = mysql(q5)
    elif sql_value[0] == '0001101000':
        result = mysql(q6)
    elif sql_value[0] == '0001110100':
        result = mysql(q7)
    elif sql_value[0] == '0001010110':
        result = mysql(q8)
    elif sql_value[0] == '0000000001':
        result = mysql(q9)
    elif sql_value[0] == '0001000010':
        result = mysql(q10)
    elif sql_value[0] == '0001110010':
        result = mysql(q11)
    else:
        result =  'Unable to Generate query!'


    return result
```
